**deal with "Q3: AGE" column**

```python
df['Q3: AGE'].describe()
```

```python
df['Q3: AGE']=pd.to_numeric(df['Q3: AGE'],errors='coerce')
```

```python
df['Q3: AGE'].unique()
```

```python
# finding outliers
Q1=df['Q3: AGE'].quantile(.25)
Q3=df['Q3: AGE'].quantile(.75)
IQR=Q3-Q1
outliers=(df['Q3: AGE']<Q1-1.5*IQR)|(df['Q3: AGE']>Q3+1.5*IQR)
df.loc[outliers,'Q3: AGE']
```

**drop age outliers**

```python
#replace outliers to nan
df['Q3: AGE']=np.where(outliers,np.nan,df['Q3: AGE'])

# fill na to mean and covert values to int 
df['Q3: AGE']=df['Q3: AGE'].fillna(df['Q3: AGE'].mean()).astype(int)
```

```python
df['Q3: AGE'].unique()
```

```python
df['Q3: AGE'].value_counts()
```

**deal with "Q4: COUNTRY" column**

```python
df['Q4: COUNTRY'].unique()
```

```python
df['Q4: COUNTRY']=df['Q4: COUNTRY'].str.strip().str.lower()
df['Q4: COUNTRY']=df['Q4: COUNTRY'].str.replace('.',"")
```

```python
# remove numbers
n=['35','46','45','32','1']
for i in df['Q4: COUNTRY']:
    if i in n:
        df['Q4: COUNTRY']=df['Q4: COUNTRY'].replace(i,np.nan)
        
#  remove not useful values     
s=['fear and loathing',"i don't know anymore",'a','subscribe to dm4uz3 on youtube','earth']
for i in df['Q4: COUNTRY']:
    if i in s:
        df['Q4: COUNTRY']=df['Q4: COUNTRY'].replace(i,np.nan)
```

```python
#usa
df['Q4: COUNTRY']=np.where((df['Q4: COUNTRY'].str.contains('sta'))|(df['Q4: COUNTRY'].str.contains('sates'))|(df['Q4: COUNTRY'].str.contains('ststes'))|(df['Q4: COUNTRY'].str.contains('us'))|(df['Q4: COUNTRY'].str.contains('u s a')),"usa",df['Q4: COUNTRY'])
#canada
df['Q4: COUNTRY']=df['Q4: COUNTRY'].replace({'canada`':'canada','can':'canada','canae':'canada'})
#united kingdom
df['Q4: COUNTRY']=df['Q4: COUNTRY'].replace({'uk':'united kingdom'})
#america
df['Q4: COUNTRY']=df['Q4: COUNTRY'].replace({"'merica":'america','merica':'america','ahemamerca':'america'})
```

```python
df['Q4: COUNTRY']=df['Q4: COUNTRY'].fillna(df['Q4: COUNTRY'].mode()[0])
df['Q4: COUNTRY']=df['Q4: COUNTRY'].str.capitalize()
```
